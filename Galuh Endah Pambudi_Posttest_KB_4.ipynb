# Data Transforming

**Normalisasi**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
dfMinMax = dfImpute.copy()
dfMinMax.head()
```

```python
minmax = MinMaxScaler() #Buat Objek scaler

dfMinMax['rating'] = minmax.fit_transform(dfMinMax[['rating']])
```

```python
print('Validasi MinMaxScaler')
print(f'Nilai Min : ', dfMinMax['rating'].min())
print(f'Nilai Max : ', dfMinMax['rating'].max())
```

# Standarisasi

```python
from sklearn.preprocessing import StandardScaler
```

```python
dfStandard = dfImpute.copy()
dfStandard.head()
```

```python
ss = StandardScaler() #Buat Objek scaler

dfStandard['rating'] = ss.fit_transform(dfStandard[['rating']])
```

```python
print('Validasi Standard Scaler')
print(f'Nilai Rata-Rata       : ', dfStandard['rating'].mean())
print(f'Nilai Standar Deviasi : ', dfStandard['rating'].std())
```
