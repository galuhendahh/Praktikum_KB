A. Supervised Learning

A1. Implementasi Support Vector Classifier

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer


data = pd.read_csv("./product_reviews_dirty.csv")


tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(data['product_name']) 

X_train, X_test, y_train, y_test = train_test_split(X, data['product_name'], test_size=0.25)  

svm = SVC()
svm.fit(X_train, y_train)

y_pred = svm.predict(X_test)

print("Classification report:")
print(classification_report(y_test, y_pred))
```

RANDOM FOREST

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer


data = pd.read_csv("./product_reviews_dirty.csv")


tfidf_vectorizer = TfidfVectorizer()
X = tfidf_vectorizer.fit_transform(data['category'])  

X_train, X_test, y_train, y_test = train_test_split(X, data['product_name'], test_size=0.25)  


rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

print("Classification report:")
print(classification_report(y_test, y_pred))
```

LINEAR REGRESSION

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


data = pd.read_csv("./product_reviews_dirty.csv")  # Ganti dengan nama file CSV Anda


X = data.drop(columns=['product_name'])  # Ganti 'target_column' dengan nama kolom target yang sesuai
y = data['product_name']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)


linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)


y_pred = linear_reg.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")


```

B. UNSPERVISED LEARNING

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, homogeneity_score, completeness_score
import matplotlib.pyplot as plt


data = pd.read_csv("product_reviews_dirty.csv")


X = data[["text", "rating"]]


k = len(data["category"].unique())


kmeans = KMeans(n_clusters=k)


kmeans.fit(X)


y_pred = kmeans.predict(X)


silhouette_score_without_labels = silhouette_score(X, y_pred)
homogeneity_score_without_labels = homogeneity_score(data["category"], y_pred)
completeness_score_without_labels = completeness_score(data["category"], y_pred)


silhouette_score_with_labels = silhouette_score(X, y_pred, labels=data["category"])
homogeneity_score_with_labels = homogeneity_score(data["category"], y_pred, labels=data["category"])
completeness_score_with_labels = completeness_score(data["category"], y_pred, labels=data["category"])


plt.scatter(X[:, 0], X[:, 1], c=y_pred, s=50, cmap="viridis")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker="*", c="black", s=100)
plt.title("K-Means Clustering")
plt.xlabel("Text")
plt.ylabel("Rating")
plt.show()

print("Silhouette score without labels:", silhouette_score_without_labels)
print("Homogeneity score without labels:", homogeneity_score_without_labels)
print("Completeness score without labels:", completeness_score_without_labels)
print()
print("Silhouette score with labels:", silhouette_score_with_labels)
print("Homogeneity score with labels:", homogeneity_score_with_labels)
print("Completeness score with labels:", completeness_score_with_labels)
```
